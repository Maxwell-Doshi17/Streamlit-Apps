```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
feature_map = {
    "1": lambda x: np.ones(len(x)),
    "x": lambda x: np.array(x),
    "x^2": lambda x: np.array(x)**2,
    "x^3": lambda x: np.array(x)**3,
    "x^0.5": lambda x: np.sqrt(x),
    "sin(x)": lambda x: np.sin(x),
    "cos(x)": lambda x: np.cos(x),
    "log(x)": lambda x: np.log(x),
}
```

```python
n = int(input("How many data points: "))
```

```python
x = []
y = []
for i in range(n):
    currentx = float(input("xval: "))
    currenty = float(input("yval: "))
    x.append(currentx)
    y.append(currenty)
x = np.array(x)
y = np.array(y)
print(x)
print(y)
```

```python
plt.scatter(x, y)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.show()
```

```python
user_input = input("Enter terms (comma separated, e.g. 1,x,x^2): ")

terms = [t.strip() for t in user_input.split(",")]
```

```python
A = np.column_stack([feature_map[t](x) for t in terms])
B = (np.array(y)).reshape(-1,1)
Ahat = A.T @ A
Bhat = A.T @ B
Xhat = np.linalg.solve(Ahat, Bhat)
```

```python
xaxis = np.linspace(0, 10, 500)
equation = np.zeros(len(xaxis))
for i, term in enumerate(terms):
    equation += Xhat[i] * feature_map[term](xaxis)
```

```python
plt.scatter(x, y)
plt.plot(xaxis, equation)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.show()
```
